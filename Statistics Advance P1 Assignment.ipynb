Ans 1: A random variable is a variable that takes on numerical values determined by the outcome of a random phenomenon. It assigns a real number to each possible outcome in a sample space.

Ans 2: There are two main types:

Discrete Random Variable: Takes on countable values (e.g., 0, 1, 2,...).

Continuous Random Variable: Takes on an uncountable range of values, typically over an interval (e.g., any real number between 0 and 1).

Ans 11: The normal distribution (Gaussian distribution) is a continuous, symmetric, bell-shaped distribution defined by its mean (μ) and standard deviation (σ).

Characteristics:

Symmetrical around the mean.

Mean = Median = Mode.

68% of data lies within 1σ of the mean.

95% within 2σ, and 99.7% within 3σ (Empirical Rule).

The curve approaches the x-axis asymptotically.

Applications:
Used in natural and social sciences, e.g., heights, test scores, measurement errors, etc.

Ans 12: The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1.

Importance:

Simplifies calculations in statistics.

Allows comparison between different normal distributions using Z-scores.

Critical in hypothesis testing and confidence intervals

Ans 13: The Central Limit Theorem states that the sampling distribution of the sample mean will tend to be approximately normally distributed, regardless of the shape of the population distribution, provided the sample size is sufficiently large (usually n ≥ 30).

Importance:

Justifies using normal distribution in inference.

Enables estimation and hypothesis testing even with non-normal population distributions.



Ans 14: The CLT explains why the normal distribution is so widely used in statistics:

Even if the population is not normally distributed, the mean of sufficiently large samples drawn from it will be approximately normal.

This allows statisticians to apply techniques based on normality (like Z-tests, t-tests) to real-world data.



Ans 15: Z statistics (Z-scores) are used in hypothesis testing when the population standard deviation is known and the sample size is large.

Applications:

Testing population means (Z-test).

Comparing proportions.

Constructing confidence intervals.

Ans 21: Independence: Observations in the sample must be independent.

Sample Size: For skewed or non-normal distributions, a larger sample (n ≥ 30) is needed.

Finite Variance: The population must have a finite standard deviation.

If these conditions are met, the sampling distribution of the mean will be approximately normal.

Ans 23: The probability distribution provides the likelihood of each possible value a random variable can take.

The expected value is computed using the probability distribution — it is essentially a weighted average of all possible outcomes, weighted by their probabilities.

Thus, the probability distribution determines the expected outcome, variance, and other properties of the random variable.
